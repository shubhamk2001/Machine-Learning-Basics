# Line Plot

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# defining the dataset
x = np.arange(0,10,0.1)
y = 2*x +5

#plotting the data-points
plt.plot (x,y)
plt.show()
```

# Customizing the line plot

```python
# defining the dataset
x = np.arange(0,10,1)
y = 2*x +5

#changing the figure-size
fig= plt.figure(figsize=(10,5))

#plotting the data-points
plt.plot (x,y,linewidth= 2.0, linestyle=':', color='g',alpha= 1,marker='o')
plt.title("line plot demo")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(['line1'],loc='best')
plt.grid('true')
plt.show()
```

```python
# Defining new Dataset
x= np.arange(0,10,1)
y1= 2*x+5
y2= 12*x+10

#create two or more plots in a fig
plt.subplot(1,2,1)

#plotting the points
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')
plt.show()
```

# How to create a bar plot

```python
# creating the dataset
data = {'apples': 20, 'Mnagoes':15, 'Lemon': 20, 'Oranges': 10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,5))

#creating the bar plot
plt.bar(names, values)
plt.show()
```

# Customizing the bar plot

```python
# creating the dataset
data = {'apples': 20, 'Mnagoes':15, 'Lemon': 20, 'Oranges': 10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,5))
#creating the bar plot
plt.barh(names, values,color='indigo')
plt.title("Fruits basket")
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```

# How to create a Scatter Plot

```python
# creating the dataset
a=[10,20,30,40,50,60,70,80]
b = [3,4,6,7,2,1,6,4]
x= [1,2,3,4,3,2,1,2]

#creating a scatter plot
plt.scatter(a,b,s=140,edgecolors='y',marker='o',alpha=0.5)
plt.scatter(a,x,s=140,edgecolors='y',marker='x',alpha=0.5)
plt.legend(['data1','data2'],loc='best')
plt.savefig('ScatterPlot.png')
plt.show()
```

# How to create a Box plot

```python
# Creating a Data
total = [34,7,5,7,85,46,5,43,6,64,56,46,33]
order = [3,52,45,4,5,24,53,45,32,5,43,53,54]
discount = [10,20,25,30,12,5,17,34,21,18,14,36,15]
data = list([total, order, discount])
# Plotting the data
plt.boxplot(data, showmeans=True)
plt.title('Boxplot')  
plt.show()
```

# Creating a Violin plot

```python
# Creating a Data
total = [34,7,5,7,85,46,5,43,6,64,56,46,33]
order = [3,52,45,4,5,24,53,45,32,5,43,53,54]
discount = [10,20,25,30,12,5,17,34,21,18,14,36,15]
data = list([total, order, discount])
# Plotting the data
plt.violinplot(data, showmeans=True, showmedians = True)
plt.title('Violin')  
plt.show()
```

# How to create a Pie Chart

```python
# preparing the dataset
label= ['Delhi', 'Mumbai','Bangaluru','Kolkata','Chennai']
flats = [3,5,7,4,1]
#creating pie chart
plt.pie(flats,labels=label, autopct='%1.1f%%',shadow=True,startangle=90, explode = (0,0.1,0,0,0))
plt.title('Flats')
plt.show()
```

# How to create a Doughnut Chart

```python
# defint the dataset
groupname = ['GroupA','GroupB',"GroupC"]
groupsize = [20,50,30]
sizecentre= [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
# Creating a doughnut chart
pie1 = plt.pie(groupsize, labels=groupname,radius= 1.5, colors=[a(0.5),b(0.2),c(0.7)])
pie2= plt.pie(sizecentre,radius=1,colors='w')
plt.show()
```
